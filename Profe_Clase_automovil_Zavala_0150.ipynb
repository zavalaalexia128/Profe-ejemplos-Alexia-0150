```python
# Clase automovil, atributos, marca, placas, precio
# funciones leerdatos(), mostrardatos()
# crear 3 objetos carro1, carro2, carro3

class Automovil:
    def __init__(self, marca, placa, precio):
        self.marca = marca
        self.placas = placa
        self.precio = precio
    def leerdatos(self):
        self.marca = input("Ingrese la marca del automovil: ")
        self.placas = input("Ingrese las placas del automovil: ")
        self.precio = input("Ingrese el precio del automovil: ")
        return self.marca, self.placas, self.precio

    def mostrardatos(self):
        print("Marca: ", self.marca)
        print("Placas: ", self.placas)
        print("Precio: ", self.precio)
        return self.marca, self.placas, self.precio
# fin de la clase automovil
# creando un objeto carro1, carro2, carro3

print("Carro Numero 1")
carro1 = Automovil( None, None, None)
carro1.leerdatos()
carro1.mostrardatos()
print(". ݁₊ ⊹ . ݁˖ . ݁")

print("Carro Numero 2")
carro2 = Automovil(None, None, None)
carro2.leerdatos()
carro2.mostrardatos()
print(". ݁₊ ⊹ . ݁˖ . ݁")

print("Carro Numero 3")
carro3 = Automovil(None, None, None)
carro3.leerdatos()
carro3.mostrardatos()
print("Software realizado por Zavala Medina Alexia Paola NL = 54")

# mostrando los datos de los objetos carro1, carro2, carro
```
